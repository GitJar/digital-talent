## 1. Deskripsi Data
Dataset yang digunakan berisi mengenai hasil ekspor mineral dari 19 Perusahaan di Indonesia pada tahun 2000-2010.
Penjelasan mengenai Variabel pada Dataset ini:

    satuan: Satuan berkenaan dengan komoditas tersebut
    tahun: Tahun
    nilai: Nilai

## 2. Statistik tentang data

```python
import pandas as pd
data_ekspor = pd.read_csv('processed-ekspor-mineral-2000-2010.csv', delimiter = ',', encoding = "utf-8")
data_ekspor
```

```python
data_ekspor.count()
```

```python
data_ekspor.columns
```

```python
data_ekspor.groupby('tahun').mean()
```

```python
data_ekspor.groupby('komoditas').mean()
```

```python
data_ekspor.groupby('nama_perusahaan').mean()
```

## 3. Visualisasi sederhana

```python
ax = data_ekspor.groupby('tahun').mean()['nilai']
ax = ax.plot(kind='bar', title ="Rata-Rata Berdasarkan Tahun", legend=True, fontsize=12)
ax.set_xlabel("Tahun", fontsize=12)
ax.set_ylabel("Nilai", fontsize=12)
```

```python
ax = data_ekspor.groupby('komoditas').mean()['nilai']
ax = ax.plot(kind='pie', title ="Rata-Rata Berdasarkan Komoditas", legend=True, labels=None)
ax.get_legend().set_bbox_to_anchor((1,1))
```

```python
ax = data_ekspor.groupby('nama_perusahaan').mean()['nilai']
ax = ax.plot(kind='pie', title ="Rata-Rata Ekspor tiap Perusahaan", legend=True, labels=None)
ax.get_legend().set_bbox_to_anchor((1,1))
```
