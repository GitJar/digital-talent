# TUGAS LIST

Diberikan sekumpulan Nilai Random antara 0 sampai dengan 100 sebanyak 250 nilai. Buatlah program untuk melakukan hal berikut:
1. Tampilkan Angka Ganjil dari ke 50 nilai tersebut
2. Tampilkan berapa banyak yang mendapatkan nilai lebih dari 70
3. Tampilkan standar deviasi dari sekumpulan nilai tersebut
4. Tampilkan nilai maksimum dan minimum
5. Tampilkan modus dari sekumpulan nilai tersebut

```python
import random

# GENERATE RANDOM
nilaiTugas = []
for i in range(250):
    nilaiTugas.append(random.randrange(1,100,1))

print(nilaiTugas)

# TAMPILKAN:
# ganjil 
# .........
nilaiGanjil = []
for i in nilaiTugas:
    if i%2 == 0:
        continue
    nilaiGanjil.append(i)
print("\nNilai Ganjil")
print(nilaiGanjil)

# nilai > 70
# .........
nilai70 = []
for i in nilaiTugas:
    if i>70:
        nilai70.append(i)
    continue
print("\nNilai > 70")
print(nilai70)

# standar deviasi
# .........
nilaikuadrat = 0
tambah = 0
jumlahnilaikuadrat = 0
banyak = len(nilaiTugas)
for i in nilaiTugas:
    tambah += i
    nilaikuadrat = i**2
    jumlahnilaikuadrat += nilaikuadrat
print("\nStandar Deviasi")
tambahkuadrat = tambah**2
variant = (((banyak*jumlahnilaikuadrat)-tambahkuadrat)/(banyak*(banyak-1)))
# (((banyak*jumlah)-pluskuadrat)/banyak*(banyak-1))
standarDev = variant**0.5
print("Jumlah Elemen = ",banyak)
print("Hasil SUM(i) = ",tambah)
print("Hasil SUM(i^2) = ",jumlahnilaikuadrat)
print("Variant = ",variant)
print("StandarDeviasi = ",standarDev)

# maksimum
# .........
nilaiMax = 0
for i in nilaiTugas:
    if i > nilaiMax:
        nilaiMax=i
    else:
        continue
print("\nNilai Max")
print(nilaiMax)

# minimum
# .........
nilaiMin = nilaiMax
for i in nilaiTugas:
    if i < nilaiMin:
        nilaiMin=i
    else:
        continue
print("\nNilai Min")
print(nilaiMin)

# modus
# .........
temp_nilaiTugas = a = nilaiTugas.copy() 
panjang = len(temp_nilaiTugas)
terkecil = temp_nilaiTugas[0]
urut = []
# Hilangkan Duplicate Value
for i in range(0,panjang):
    j = i + 1
    while j < len(a):
        if a[i] == a[j]:
            del a[j]
        else:
            j += 1
# Sorting
i=0
while len(temp_nilaiTugas) > 0:
    if temp_nilaiTugas[i] < terkecil:
        terkecil = temp_nilaiTugas[i]
    i += 1
    if i == len(temp_nilaiTugas):
        urut.append(terkecil)
        temp_nilaiTugas.remove(terkecil)
        if temp_nilaiTugas:
            terkecil = temp_nilaiTugas[0]
        i = 0
        
maksFrequency = 0
modusData = []
dictionaryNilai = {}
for nilai in urut:
    freq = nilaiTugas.count(nilai)
    dictionaryNilai[nilai] = freq
    if freq>maksFrequency:
        maksFrequency = freq

for nilai,jumlah in dictionaryNilai.items():
    if jumlah == maksFrequency:
        modusData.append(nilai)
        
print('\nModus')
print("List ",urut)
print('\nData yang paling sering muncul =',modusData)
print('Dengan Jumlah = ',maksFrequency)
```
